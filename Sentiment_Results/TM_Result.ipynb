```python
cna = pd.read_excel("cna.xlsx")
cna_list = cna.values.tolist()

bbc = pd.read_excel("bbc.xlsx")
bbc_list = bbc.values.tolist()

cnbc = pd.read_excel("cnbc.xlsx")
cnbc_list = cnbc.values.tolist()

nytimes = pd.read_excel("nytimes.xlsx")
nytimes_list = nytimes.values.tolist()

ukrinform = pd.read_excel("ukrinform.xlsx")
ukrinform_list = ukrinform.values.tolist()

tass = pd.read_excel("tass.xlsx")
tass_list = tass.values.tolist()
```

```python
def positive_bold(val):
    bold = 'bold' if val == "Positive" or val =="Negative" or "Neutral" else ''
    return 'font-weight: %s' % bold

def positive_green(val):
    color = 'green' if val == "Positive" else ''
    return 'color:' + color  

def negative_red(val):
    color = 'red' if val == "Negative" else ''
    return 'color:' + color  
```
